```python
import pandas as pd
import numpy as np
import io
```

```python
#we are taking only five column for on hot encoding
df = df[['X1','X2','X3','X4','X5','X6']]
```

```python
# lets check how many catogries are present are in each column
for col in df.columns:
  print(col,':', len(df[col].unique()))
```

```python
pd.get_dummies(df,drop_first = True)
```

```python
## After performing one hot encoding by using get dummies, the no columns become 117. if there are lots of categorical variable are present then# 
#this leads to tye curse of dimensionality

# so we take only 10 most frequent cat variable and rest will be ignored.
```

```python
# lets take 10 most frequent cat. variable pesent in any columns

values_count = df['X2'].value_counts().sort_values(ascending = False).head(10)
```

```python
top_10_value = [x for x in values_count.index ]
```

```python
top_10_value
```

```python
## now covert these variable into 0 and 1

for label in top_10_value:
  df[label] = np.where(df['X2'] == label,1,0)
df[['X2']+top_10_value]
```

```python
## now aplly these concept in each and every columns.

def one_hot_encode(data,variable,top_x):
  for label in top_x:
    df[variable+"-"+label] = np.where(data[variable] ==label,1,0)
```

```python
df = df[['X1','X2','X3','X4','X5','X6']]
```

```python
one_hot_encode(df,'X1',top_10_value)
df.head()
```
